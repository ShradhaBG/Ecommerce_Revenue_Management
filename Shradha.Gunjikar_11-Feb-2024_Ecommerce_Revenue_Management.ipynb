    The E-Commerce company has the following rule for computing their own revenue: </font>
    
####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font>

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue.

### Ans 5

Revenue_of_E-commerce

```python
# Defining a function using def command to calculate revenue according to the above specified conditions
def Revenue_of_e_commerce(discounted_price):
    if discounted_price > 600:
        return 0.25 * discounted_price
    elif 350 < discounted_price <= 600:
        return 0.15 * discounted_price
    elif 100 < discounted_price <= 350:
        return 0.10 * discounted_price
    else:
        return 0.05 * discounted_price

 #using apply() function to create a new column
ECom_Data['Total_revenue'] = ECom_Data['discounted_price'].apply(Revenue_of_e_commerce)

total_revenue = ECom_Data['Total_revenue'].sum()

print('Total Revenue of E-Commerce company:',total_revenue)
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


### Ans 6

```python
#Brand_revenue for each product
ECom_Data['brand_revenue'] = ECom_Data['discounted_price'] - ECom_Data['Total_revenue']

# Calculating Total BrandRevenue for each brand
total_brand_revenue = ECom_Data.groupby('Brand')['brand_revenue'].sum()

# Listing the top 10 Brands with maximum revenue in descending order
top_10_brands = total_brand_revenue.nlargest(10)

print("Top 10 Brands with maximum revenue:",top_10_brands)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks)

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import calendar


# Converting 'Order_Date' to datetime format
ECom_Data['Order_Date'] = pd.to_datetime(ECom_Data['Order_Date'], format='%d/%m/%Y')#using format parameters to get rid of any waranings it is telling pandas to interpret the dates using the 'DD/MM/YYYY' format

# Extracting Month and Year from Order_Date
ECom_Data['Month'] = ECom_Data['Order_Date'].dt.month
ECom_Data['Year'] = ECom_Data['Order_Date'].dt.year

# Filling missing values in 'discounted_price' and 'Total_revenue' columns with 0
ECom_Data['discounted_price'].fillna(0, inplace=True) #inorder to avoid value error using fillna() method(replaces the null value with a specifies value)
ECom_Data['Total_revenue'].fillna(0, inplace=True)

# Correcting the calculation of revenue
ECom_Data['revenue'] = ECom_Data['discounted_price'] - ECom_Data['Total_revenue']

# Creating a lineplot for monthly Revenue for each region with markers
plt.figure(figsize=(12, 6))
sns.lineplot(data=ECom_Data, x='Month', y='revenue', hue='Region', style='Region', markers=True, err_style=None)

# Extracting unique months and years from the data
unique_months_years = ECom_Data[['Month', 'Year']].drop_duplicates().sort_values(by=['Year', 'Month'])

# Setting up x-axis ticks and labels with rotation
plt.xticks(range(len(unique_months_years)),
           [f"{calendar.month_abbr[m]} {y}" for m, y in zip(unique_months_years['Month'], unique_months_years['Year'])],
           rotation=45)
plt.title('Monthly Revenue of E-Commerce Company by Region')
plt.xlabel('Month and Year')
plt.ylabel('Revenue')
plt.legend(title='Region')
plt.show()
```

### Ans 9 b)

```python
# the best and the worst performing months for each region
best_per_months = ECom_Data.groupby(['Region', 'Year', 'Month'])['revenue'].sum().idxmax()
worst_per_months = ECom_Data.groupby(['Region', 'Year', 'Month'])['revenue'].sum().idxmin()

print("Best Performing Months:",best_per_months)


print("Worst Performing Months:",worst_per_months)
```
